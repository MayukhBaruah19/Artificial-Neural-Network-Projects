```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancre_dataset=sklearn.datasets.load_breast_cancer()
print(breast_cancre_dataset)
```

```python
data=pd.DataFrame(breast_cancre_dataset.data,columns=breast_cancre_dataset.feature_names)
```

```python
data['label'] = breast_cancre_dataset.target
data.head()
```

```python
X=data.drop('label',axis=1)
y=data['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

### Implimentation of ANN

```python
# importing tensorflow and Keras
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape[1],
```

```python
# setting up the layers of Neural Network
model=Sequential(
    [
        Dense(30,activation='relu',input_shape=(X_train.shape[1],)), #HL1 Connwcted wit input layer
        Dense(20,activation='relu'),#HL2
        Dense(2,activation='sigmoid')
    ]
    
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#Tranning the neural network 
history=model.fit(X_train_scaled,y_train,validation_data=(X_test,y_test),epochs=10,validation_split=0.1)
```

```python
history.history
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(accuracy)
```
